```python
def set_distance(head_df):
    for i in range(len(head_df)):
        start = head_df.iloc[i]['start_station_latitude'], head_df.iloc[i]['start_station_longitude']
        end = head_df.iloc[i]['end_station_latitude'], head_df.iloc[i]['end_station_longitude']
        head_df['distance'].iloc[i] = haversine(start, end)
```

```python
my_time = df['duration']
factors = (60, 1, 1/60)

t1 = sum(i*j for i, j in zip(map(int, my_time.split(':')), factors))
t2 = sum(map(mul, map(int, my_time.split(':')), factors))

print(t1) 

assert t1 == t2
```

```python
df['minutes'] = timedelta(minutes = df.iloc[0]['duration'])
```

```python
def get_seconds():
    count=0
    for i in df.iloc[count]['duration']:
        # split in hh, mm, ss
        hh, mm, ss = i.split(':')
        df.iloc[count]['time_in_minutes'] = mm + ss/60
        count += 1
```
