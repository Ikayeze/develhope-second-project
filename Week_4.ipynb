```python
def sliding_windows(df):
    liste = [[]]
    count = 0
    for i in range(len(df)):
        if i % 2 == 0:
            try:
                liste[count].append(df.iloc[i:i+2])
                count += 1
                liste.append([])
            except:
                liste[count].append(df.iloc[i:])
        else:
            pass
    
```

```python
sliding_windows(df)
```

```python
def sliding_window_2(df, size):
    mylist = []
    if len(df) <= size:
        return df
    for i in range(len(df) - size + 1):
        try:
            mylist.append([df[i:i+size]])
        except:
            mylist.append([df[i:]])
```
