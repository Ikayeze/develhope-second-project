```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/mkaya/Downloads/googleplaystore.csv")
```

```python
df.sort_values("Installs", ascending = False).head()
```

```python
df.loc[10472] = ["Life Made WI-Fi Touchscreen Photo Frame", np.nan, 1.9, 19.0, "3.0M", "1000+", "Free", 0, "Everyone", np.nan, "February 11,2018", "1.0.19", "4.0 and up"]
```

```python
df.sort_values("Installs", ascending = False).head()
```

```python
display(df["Category"].unique())
```

```python
liste = []
for i in df["Installs"]:
    i = i.replace("+","")
    i = i.replace(",", "")
    liste.append(i)
```

```python
df["Installs"] = liste
```

```python
df["Installs"] = df["Installs"].astype("float")
```

```python
installs_by_category = df.groupby("Category")["Installs"].sum().to_frame()
```

```python
plt.figure(figsize=(20,10))
plt.title("Total Number of Installs by Category")
plt.xticks(rotation = 90)
plt.bar(installs_by_category.index, installs_by_category["Installs"])
```

```python
liste = []
for i in df["Price"]:
    i = str(i).replace("$", "")
    liste.append(i)
```

```python
df["Price"] = liste
```

```python
df["Price"] = df["Price"].astype("float")
```

```python
df.info()
```

```python
df_paid = df[df["Price"] > 0].copy()
```

```python
df_paid.groupby("Category").sum().Price.to_frame()
```

```python
df_paid[df_paid["Category"] == "ART_AND_DESIGN"]
```

```python
plt.figure(figsize=(10,5))
plt.xlabel("Name of Categories")
plt.ylabel("Total Pay For Each Category")
plt.xticks(rotation = 90)
plt.bar(df_paid.groupby("Category").sum().index, df_paid.groupby("Category").sum().Price)
```

```python
df_group = df_paid.groupby("Category").sum()
```

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation = 90)
plt.bar(df_group.index, df_group["Installs"] * df_group["Price"])
plt.show()
```
